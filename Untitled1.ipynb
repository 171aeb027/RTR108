\begin{equation}
cos({x})*cos({x})=1-\sum_{k=1}^{k=500}\frac{(-1)^{k+1}\cdot x^{2*k}\cdot 2^{2*k-1}}{(2\cdot k) !}==(1-\frac{(-1)^{1+1}\cdot x^{2*1}\cdot 2^{2*1-1}}{2\cdot 1 !}) + (1-\frac{(-1)^{2+1}\cdot x^{2*2}\cdot 2^{2*2-1}}{2\cdot 2 !})+(1-\frac{(-1)^{3+1}\cdot x^{2*3}\cdot 2^{2*3-1}}{2\cdot 3 !})+(1-\frac{(-1)^{4+1}\cdot x^{4*1}\cdot 2^{2*4-1}}{2\cdot 4 !})
\end{equation}

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 50.0, 0.01)
s = np.cos(t)*np.cos(t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)', title='cos(t)*cos(t)')
ax.grid()

fig.savefig("test.png")
plt.show()
```
